```python
# Loading the Earthquake dataset
import pandas as pd

# Load CSV file into a DataFrame
file_path = 'earthquakes_2023_global.csv'
df1 = pd.read_csv(file_path)
df1
```

```python
# Removing missing values
df1 = df1.dropna()
df1 = df1.reset_index(drop=True)
```

```python
# Creating Area, last location in the Place column
df1['area'] = df1['place'].str.split(',').str[-1].str.strip()
df1
```

```python
# Loading the Tectonic Plate Dataset
import pandas as pd

file_path2 = 'all.csv'
df2 = pd.read_csv(file_path2)
df2
```

```python
# Removing missing values
df2 = df2.dropna()

df2 = df2.reset_index(drop=True)
```

```python
# Creating the distance column

import pandas as pd
from scipy.spatial import cKDTree
from math import radians, sin, cos, sqrt, atan2

# Function to calculate Haversine distance
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Radius of Earth in kilometers

    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)

    a = sin(dlat / 2) ** 2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2) ** 2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Convert latitude and longitude to radians
df1['latitude_rad'] = df1['latitude'].apply(radians)
df1['longitude_rad'] = df1['longitude'].apply(radians)
df2['latitude_rad'] = df2['lat'].apply(radians)
df2['longitude_rad'] = df2['lon'].apply(radians)

# Create a KDTree for tectonic plates
tree = cKDTree(df2[['latitude_rad', 'longitude_rad']])

# Query the KDTree for each earthquake in df1
distances, indices = tree.query(df1[['latitude_rad', 'longitude_rad']], k=1)

# Update df1 with the minimum distance, corresponding plate location, and plate name
df1['min_distance_to_tectonic_plate'] = distances
df1[['plate_latitude', 'plate_longitude']] = df2[['lat', 'lon']].iloc[indices].values
df1['plate_name'] = df2['plate'].iloc[indices].values

# Convert distances back to kilometers
df1['min_distance_to_tectonic_plate'] *= 6371

# Drop the temporary columns
df1.drop(['latitude_rad', 'longitude_rad'], axis=1, inplace=True)
```

```python
# Saving the updated dataset as a CSV file.
output_file_path3 = 'Eartquakes_w_plates.csv'
df1.to_csv(output_file_path3, index=False)

print("Modifications saved successfully.")
```
